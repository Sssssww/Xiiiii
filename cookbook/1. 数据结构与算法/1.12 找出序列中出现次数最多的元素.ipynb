```python
from collections import Counter

items = [
"我","如","果","爱","你","绝","不","像","攀","援","的","凌","霄","花","借","你","的","高","枝","来","炫","耀","自",
    "己","我","如","果","爱","你","绝","不","学","痴","情","的","鸟","儿","向","绿","荫","重","复","单","调","的","歌","曲"
]
items_counter = Counter(items)
top_three = items_counter.most_common(2)
print(top_three)
```

```python
print(items_counter["我"])
print(items_counter["绝"])
```

```python
# 如果想增加计数，有两种方法
new_items = ["也", "不", "止", "像", "源", "泉", "常", "年", "送", "来", "清", "凉", "的", "慰", "藉"]
for item in new_items:
    items.append(item)
print(items)
```

```python
# 也可以使用update()方法  
# 此方法未改变items 改变了items_counter而上面方法改变的items

new_items = ["也", "不", "止", "像", "源", "泉", "常", "年", "送", "来", "清", "凉", "的", "慰", "藉"]
items = [
"我","如","果","爱","你","绝","不","像","攀","援","的","凌","霄","花","借","你","的","高","枝","来","炫","耀","自",
    "己","我","如","果","爱","你","绝","不","学","痴","情","的","鸟","儿","向","绿","荫","重","复","单","调","的","歌","曲"
]
print(items)
print("="*20)
items_counter.update(new_items)
print(items_counter)
print(items)
```

```python
# couter 可以进行各种数学运算操作
new_items = ["也", "不", "止", "像", "源", "泉", "常", "年", "送", "来", "清", "凉", "的", "慰", "藉"]
items = [
"我","如","果","爱","你","绝","不","像","攀","援","的","凌","霄","花","借","你","的","高","枝","来","炫","耀","自",
    "己","我","如","果","爱","你","绝","不","学","痴","情","的","鸟","儿","向","绿","荫","重","复","单","调","的","歌","曲"
]
a = Counter(new_items)
b = Counter(items)
c = a + b
print(c)
d = c - a
print(d)
```
